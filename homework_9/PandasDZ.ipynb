# Задание 1
## Скачайте с сайта grouplens.org...movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
merged = ratings.merge(movies, on = 'movieId', how = 'left')
merged[['movieId', 'rating', 'title']].head()
```

```python
max_rat = ratings['rating'].max()
filt = merged[(merged ['rating'] == max_rat)]
best = filt.groupby(['movieId' , 'title'])['movieId'].count()
```

```python
best.sort_values(ascending=False).head()
```

# Задание 2
## По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
filt = power[(power['country'] == 'Latvia') | (power['country'] == 'Lithuania') | (power['country'] == 'Estonia') ]
filt.head()
```

```python
quantity = filt[(filt['quantity'] > 0)]
```

```python
category = quantity[(quantity['category'] == 4) | (quantity['category'] == 12) | (quantity['category'] == 21)]
```

```python
year = category[(category['year'] < 2010) & (category['year'] > 2005)]
```

```python
print('Cуммарное потребление: {}'.format(year['quantity'].sum()))
```
